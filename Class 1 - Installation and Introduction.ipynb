2. <b>Run some basic commands</b>

- <b>math</b>
    - add: 1`+`2
    - substract: 1`-`2
    - multiply: 1`*`2 
    - divide: 1`/`2
    - power: 2`**`2

- <b>string</b>
    - echo: 'hi'
    - extend:'hi'`+`'!'
    - duplicate: 'a'`*`20

- <b>comparison</b>: also called <b>logical conditions</b>, because its return value is logical boolean value, i.e. True or False.
    - equal: 1 `is` 0 or 1 `==` 0
    - not equal: 1 `is not` 0 or 1 `!=` 0
    - less than: 1 `<` 0
    - less than or equal: 1 `<=` 0
    - greater than: 1 `>` 0
    - greater than or equal: 1 `>=` 0

- <b>boolean operation</b>:
    - and: True `and` False, (1`>`0) and (1`==`0)
    - or: True `or` False, (1`>`0) or (1`==`0)
    - not:`not` True, not (1`==`0)

- <b>variable</b>:
    - definition: Variables point to values. They are the nicknames of values, designated by user.
    - assign values to variable: a `=` 1, here "a" is the variable name and 1 is its value.
    - increase variables by 1 unit: a `=` a + 1 or a `+=` 1
    - decrease variables by 1 unit: a `=` a - 1 or a `-=` 1

<div class="alert alert-block alert-warning">
**Warning**: Python is a case-sensitive language. So, if you try "1 IS 0", it will return error message.
</div>

<div class="alert alert-block alert-info">
**Tip:** To quickly copy last line, you can press <kbd></kbd>. To copy the second last lin, you can press <kbd></kbd> twice. And so on.

4. <b>Summary</b>
<p>Generally speaking, every operating system has equipped with Linux-like **built-in CLIs** by default. They can be instantly switched to Python environment by simply one command, i.e. "py" or "py3".</p>
<p>Besides, a **Python-specific CLI** is provided after installation. You could find it by searching its name in Start Menu or Spotlight. Different from built-in CLIs, it is running under python environment by default so it doesn't require extra command to turn into this way.</p>
<p>However, the biggest weakness of CLIs is that you can only write commands **line by line**, which is inefficient and even disruptive to integrative thinking. To overcome this, we will introduce two alternatives to CLIs, namely **Jupyter Notebook** and **Sublime Text (optional)**.</p>

<h3>C. Alternatives to CLIs</h3>

<img src="img/week1-jupyter.png" width="120" align='left'><br><p>Other than CLIs, we can choose to use external softwares to run Python environment.</p><p>Here we will first learn an <font style="color:red">interactive</font> python editor called **Jupyter Notebook**, which has been widely adopted as a norm in Silicon Valley.</p>

1. **install jupyter via pip**
    - **pip**:
        - pip is an installation assistant library, which has been installed along with Python.
        - usage: you can use pip by typing `pip3 install [library name]` in system built-in CLIs (Terminal/PowerShell/Command Prompt).
        - example: to install jupyter notebook, you should type `pip3 install jupyter`

<div class="alert alert-block alert-info">
**Tip**: Here we need to use "pip3" instead of "pip", because by specifying "3" we can order pip to install a library that supports Python 3.X.

2. **run jupyter notebook**
    - type "jupyter notebook" in system built-in CLIs
    - it will automatically open a new page in your default browser
    - the page provides a view of current folder

<img src='./img/week1-jupyter screenshot.png' width="600" align='left'>

<div class="alert alert-block alert-info">
